## Q1 Number of matches played in citys

```python
df1['city'].sort_values().unique()   #total city
```

```python
df1['city'] = df1['city'].replace(['Chandigarh'],'Mohali') #replace 'chandigarh' to mohali
```

```python
city_name = df1.city.value_counts()   # count the number of matches city have
city_name = pd.DataFrame(city_name)   # convert it into the dataframe
city_name = city_name.reset_index()   # rest the index 
city_name.columns = ['city','number of matchs']  # give new column name
city_name = city_name.sort_values(by=['city'])   # sort by city names
```

```python
fig = plt.gcf()
fig.set_size_inches(18,10)
plt.bar(city_name['city'],city_name['number of matchs'],color =['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25','#E82865'] ,alpha=.8)
xlocs, xlabs = plt.xticks()
plt.ylabel("Number of matchs")
plt.xlabel("citys name")
plt.title("Number of matchs played in citys")
plt.xticks(rotation=90)
#plt.xticks(xlocs, xlabs)
for i, v in enumerate(city_name['number of matchs']):      # picked from stackoverflow(for showing number of matches)
    plt.text(xlocs[i] - 0.25, v + 0.01, str(v))
plt.show()
```

## Q2 Number of win by individual team

```python
df1['winner'].sort_values().unique()
```

```python
df1['winner'] = df1['winner'].replace(['Rising Pune Supergiants'],'Rising Pune Supergiant')
```

```python
match_wins = df1.winner.value_counts()
match_wins = pd.DataFrame(match_wins)
match_wins = match_wins.reset_index()
match_wins.columns = ['Team Name','Matchs win']
match_wins = match_wins.sort_values(by='Team Name')
```

```python
sns.set_style("whitegrid")
fig = plt.gcf()
fig.set_size_inches(10,5)       #set the area for bar plot
plt.bar(match_wins['Team Name'],match_wins['Matchs win'],color =['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25'] )
xlocs, xlabs = plt.xticks()
plt.ylabel("Number of matchs win by individual")
plt.xlabel("Team Name",fontsize = 20)
plt.title("Number of Win",fontsize = 25)
plt.xticks(rotation=90)
#plt.xticks(xlocs, xlabs)
for i, v in enumerate(match_wins['Matchs win']):      # picked from stackoverflow(for showing number of matches)
    plt.text(xlocs[i] - 0.25, v + 0.01, str(v))
plt.show()
```

## Q4. To 10 player with most MOM 

```python
mom = df1.player_of_match.value_counts()
mom = pd.DataFrame(mom)
mom = mom.reset_index()
mom.columns = ['Player_name','MOM']
mom.head()
```

```python
#sns.set_style("whitegrid")
sns.set_style("darkgrid")
fig = plt.gcf()
fig.set_size_inches(10,7)
plt.xticks(rotation=90,fontsize =0)
plt.bar(mom['Player_name'][:15],mom['MOM'][:15],color=[
                                        '#294A73','#D4483B','#FD511F','#158EA6'
                                        ,'#E82865','#005DB7','#C23E25'],alpha=.7)   
        
count =0;
for x in mom['MOM'][:15]:    #this is used for showing player name inside bar
    plt.text(count,5,mom['Player_name'][count]+':'+str(x),rotation = 90,color='black')
    count +=1
plt.title("Most man of the match")
plt.xlabel("Player",fontsize=20)
plt.ylabel("No. of time of won")
plt.tight_layout()
plt.show()
```
